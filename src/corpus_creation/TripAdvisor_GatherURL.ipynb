```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
import requests
import csv
import time
import re
from selenium import webdriver
import os
```

```python
# Gathering 3,000 unique URLs for different experiences

urls = set()

# iterate through location 240 - 440 on Tripadvisor's website
for i in range(240,440):
    print(f'processing location {i}...')
    if len(urls) < 3000: # keep collecting URLs for different experiences until we have collected 3000 unique URLs
        url = "https://www.tripadvisor.com/Attractions-g147" + str(i) + "-Activities-c42"
        if i % 10 == 0:
            print(f'Processed {i-239} location(s) and obtained {len(urls)} total url(s)')
        try:
            #try to get the urls for the top 30 experiences for a location if available
            soup = BeautifulSoup(urlopen(url))
            for node in soup.find_all("div", {"class": "_3TW1AwDl _2mmsZnD4 _3dIDcq-c _3ZgRmtgg"}):
                s = re.search('href="(.*)" target', str(node.find("a")))
                if s != None:
                    urls.add('https://www.tripadvisor.com'+s.group(1))
            time.sleep(1)
            try:
                #try to get the urls for the top 30-60 experiences for a location if available
                url = "https://www.tripadvisor.com/Attractions-g147" + str(i) + "-Activities-c42-oa30"
                soup = BeautifulSoup(urlopen(url))
                for node in soup.find_all("div", {"class": "_3TW1AwDl _2mmsZnD4 _3dIDcq-c _3ZgRmtgg"}):
                    s = re.search('href="(.*)" target', str(node.find("a")))
                    if s != None:
                        urls.add('https://www.tripadvisor.com'+s.group(1))
                time.sleep(1)
            except:
                pass
            try:
                #try to get the top 60-90 experiences for a location if available
                url = "https://www.tripadvisor.com/Attractions-g147" + str(i) + "-Activities-c42-oa60"
                soup = BeautifulSoup(urlopen(url))
                for node in soup.find_all("div", {"class": "_3TW1AwDl _2mmsZnD4 _3dIDcq-c _3ZgRmtgg"}):
                    s = re.search('href="(.*)" target', str(node.find("a")))
                    if s != None:
                        urls.add('https://www.tripadvisor.com'+s.group(1))
                time.sleep(1)
            except:
                pass
            try:
                #try to get the urls for the top 90-120 experiences for a location if available
                url = "https://www.tripadvisor.com/Attractions-g147" + str(i) + "-Activities-c42-oa90"
                soup = BeautifulSoup(urlopen(url))
                for node in soup.find_all("div", {"class": "_3TW1AwDl _2mmsZnD4 _3dIDcq-c _3ZgRmtgg"}):
                    s = re.search('href="(.*)" target', str(node.find("a")))
                    if s != None:
                        urls.add('https://www.tripadvisor.com'+s.group(1))
                time.sleep(1)
            except:
                pass
            try:
                #try to get the urls for the top 120-150 experiences for a location if available
                url = "https://www.tripadvisor.com/Attractions-g147" + str(i) + "-Activities-c42-oa120"
                soup = BeautifulSoup(urlopen(url))
                for node in soup.find_all("div", {"class": "_3TW1AwDl _2mmsZnD4 _3dIDcq-c _3ZgRmtgg"}):
                    s = re.search('href="(.*)" target', str(node.find("a")))
                    if s != None:
                        urls.add('https://www.tripadvisor.com'+s.group(1))
                time.sleep(1)
            except:
                pass
        except:
            pass
                
        time.sleep(1)
```

```python
# storing all urls inside a .csv file
save_path = "../../data/tripadvisor_urls.csv"

count = 0
with open(save_path, 'w', newline='', encoding="utf-8") as csv_file:
    csvWriter = csv.writer(csv_file)
    csvWriter.writerow(['url']) 
    for url in urls:
        csvWriter.writerow([url])
        count += 1
```
