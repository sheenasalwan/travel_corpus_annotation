```python
def get_corpus_stats(corpus):
    word_list = []
    for index,document in enumerate(corpus):
        word_list.append("".join((" ".join(document[1:]))))
        
    corpus_str = " ".join(word_list)
    corpus_word_tokenized = word_tokenize(corpus_str)
    corpus_sent_tokenized = sent_tokenize(corpus_str)
    
    return len(corpus_word_tokenized),len(set(corpus_word_tokenized)),len(corpus_sent_tokenized),index, " ".join(corpus_word_tokenized)
```

```python
import pandas as pd
import numpy as np
import nltk
from nltk import word_tokenize,sent_tokenize

corpus = pd.read_csv("../../data/final_corpus.csv", usecols = ["Document ID","Overview","Review1","Review2","Review3","Review4","Review5"])
corpus = corpus.replace(np.nan, '', regex=True)
text_list = corpus.values.tolist()

num_tokens, num_types, num_sents, num_docs, tokenised_corpus = get_corpus_stats(text_list)
print("====Corpus Stats for corpus created from airbnb and tripadvisor overview and reviews====")
print("Total number of documents: ",num_docs)
print("Total number of sentences: ",num_sents)
print("Total number of tokens: ",num_tokens)
print("Total number of types: ",num_types)
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
stopwords = set(STOPWORDS)
stopwords.add('will')
stopwords.add('thing')

wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(tokenised_corpus) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```
